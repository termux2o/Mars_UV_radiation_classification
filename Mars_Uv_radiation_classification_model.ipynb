**Over sampling because under sampling valuable data loss**

```python

x=d.drop(['earth_date_time','mars_date_time','sol_number','sunrise','sunset',"UV_Radiation","weather",'radiation_categories',"humidity(%)"],axis=1)
y=d['radiation_categories']#dpendent var
```

```python
x.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x_res,y_res=SMOTE().fit_resample(x,y)
```

```python
x_res.shape
```

```python
y_res.shape
```

```python
y_res.value_counts()
```

```python
#x_res,y_res means balanced datset
from sklearn.model_selection import train_test_split
x_t2,x_tst2,y_t2,y_tst2=train_test_split(x_res,y_res,test_size=0.20,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
ft=RandomForestClassifier()
ft.fit(x_t2,y_t2)
```

```python
new1_prediction=ft.predict(x_tst2)
new1_prediction
```

```python
precision_score(y_tst2,new1_prediction,average='macro')
```

```python
accuracy_score(y_tst2,new1_prediction)
```

```python
f1_score(y_tst2,new1_prediction,average='macro')
```

```python
recall_score(y_tst2,new1_prediction,average='macro')
```
